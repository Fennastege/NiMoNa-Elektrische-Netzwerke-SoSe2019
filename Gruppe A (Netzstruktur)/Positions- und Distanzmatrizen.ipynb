```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

from scipy import optimize

import locale
locale.setlocale(locale.LC_NUMERIC,"german")
plt.rcdefaults()
plt.rcParams['axes.formatter.use_locale']=True

sb.set_style("darkgrid")

from scipy.signal import argrelextrema
```

```python
# Anzahl der Oszillatoren/ Größe des Systems festlegen
N = 15

# Normalverteilte Positionen
Xpos = np.random.normal(75, 30, N)
Ypos = np.random.normal(75, 30, N)
posmat = np.asarray([Xpos, Ypos])

#print(posmat)
```

```python
# Plot zur Veranschaulichung

%matplotlib notebook
sb.set_style("darkgrid")

plt.plot(posmat[0], posmat[1], marker="x", linestyle="", zorder=5 ,markersize=8, color="black", label="Oszillatoren")


plt.xlabel("Position X", fontsize=14)
plt.ylabel("Position Y", fontsize=14)
plt.legend(frameon=True, facecolor="lightgrey", fontsize=12)
```

```python
# Berechnung der Distanz zwischen zwei beliebigen Oszillatoren

Dist = np.ndarray(shape=(N,N))

for i in range(N):
    for j in range(N):
        Dist[i,j] = np.sqrt((posmat[0,j] - posmat[0,i])**2 + (posmat[1,j] - posmat[1,i])**2)
        
print(Dist)
```
